```python
import pandas as pd 
```

```python
data = pd.read_excel('/home/simplon/Téléchargements/chomage_zone.xlsx',sheet_name = 'txcho_ze',skiprows=5,header=0)
print(data)
```

```python
data_grp = data.groupby('LIBREG', as_index=False).mean()
data_grp
```

```python
data_grp.plot(x='LIBREG',y='2020-T1', kind='bar',grid=True, figsize=(15,15),title='taux de chômage du premier trimestre 2020 par région', fontsize=15)
```

```python

data2=data.iloc[:,range(3,73)]
data2

```

```python
data_grp2 = data2.groupby('LIBREG').mean()
data_grp2
```

```python
data_transpose = data_grp2.transpose()
data_transpose
```

```python
data_transpose = data_transpose.reset_index()

data_transpose
```

```python
data_transpose = data_transpose.rename(columns={'index':'période'})
data_transpose
```

```python
data_transpose.plot(x='période',y="Provence-Alpes-Côte d'Azur",grid=True, figsize=(15,15),title='taux de chômage du premier trimestre 2020 de la région paca', fontsize=15)
```
